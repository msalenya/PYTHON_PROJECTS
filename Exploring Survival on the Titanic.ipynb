```python
#Dropping Columns

train.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin', 'Parch'], axis=1, inplace=True)

test.drop(columns= ['PassengerId', 'Name', 'Ticket', 'Cabin', 'Parch'], axis=1, inplace= True)
```

```python
#Replacing the Age null values with the median for the Train Data

train['Age'].fillna(train['Age'].median(), inplace=True)
```

```python
#Replacing the Embacked null values with the mode for the Train Data

train['Embarked'].fillna(train['Embarked'].mode()[0], inplace=True)
```

```python
#Calculating the Median Age for the Test Data

test['Age'].median()
```

```python
#Calculating median for fare in Test Data

test['Fare'].median()
```

```python
#Replacing the Fare null values with the median for the Test Data

test['Fare'].fillna(test['Fare'].median(), inplace=True)
```

```python
#Encoding Categorical Variables

train= pd.get_dummies(train, columns=["Pclass", "Sex", "Embarked"])

test= pd.get_dummies(test, columns=["Pclass", "Sex", "Embarked"])
```

```python
#Selecting Independent and Dependent variables

X = train.drop("Survived", axis=1)
Y = train.Survived
```

```python
#Splitting the dataset into training and test data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
#Standardizing data from train and test dataset

ss= StandardScaler()

features= ['Age', 'SibSp', 'Fare']

X_train[features] = ss.fit_transform(X_train[features])
X_test[features] = ss.transform(X_test[features])
```

```python
#Model Selection  (LOGISTIC REGRESSION)

LogReg = LogisticRegression()

LogReg.fit(X_train, Y_train)
y_pred = LogReg.predict(X_test)


#Printing the mean squared error for the test data for Logistic Regression
mse = mean_squared_error(Y_test, y_pred)
rmse = mse ** 0.5
rmse
```

```python
#Model Selection  (RANDOM FOREST)


rfc = RandomForestClassifier(n_estimators=100, random_state=42)

rfc.fit(X_train, Y_train)

y_predict= rfc.predict(X_test)
```

```python
#Printing the mean squared error for the test data for Random Forest

mse = mean_squared_error(Y_test, y_predict)
rmse = mse ** 0.5
rmse
```
